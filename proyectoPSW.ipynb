```
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```
l= input("Ingrese la id: ")
n= "/anythingg?tab=tags&sort=votes&page=1"
w = "https://es.stackoverflow.com/users/"
url = w+l+str(n)
page = requests.get (url)
soup = BeautifulSoup(page.content, 'html.parser')

gg= soup.find_all('a', class_= 'post-tag')


etiquetas = list ()

for i in gg:
    etiquetas.append(i.text)

etiquetas.pop(0)

gt= soup.find_all('div', class_= 'answer-votes')

count = 1
numeros = list ()

for i in gt:
    numeros.append(i.text)


datos = {'Nombre': etiquetas, 'Puntos': numeros}
df = pd.DataFrame({'Nombre': etiquetas, 'Puntos': numeros})
df.to_csv('Gg.csv', index=False)
#print (df)
df.to_csv('Gg.csv', index=False)

text = pd.read_csv("Gg.csv")

df[['Nombre', 'Puntos']].head()

textvar = ' '.join(df['Nombre'].tolist())

wordcloud = WordCloud(background_color='black', width=3000,height=2000,stopwords=STOPWORDS).generate(textvar)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

print (df)
```
